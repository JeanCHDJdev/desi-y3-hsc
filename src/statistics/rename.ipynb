## `rename.ipynb`

For all renaming purposes for single specific runs.

```python
import numpy as np 
from pathlib import Path
```

```python
# we want to go from b2x1 to b2x3 with all file paths
outroot = Path('outputs/v8_symexpr_40/cross')
tracers = ['BGS_ANY', 'ELGnotqso', 'QSO', 'LRG']
for t in tracers:
    tracerroot = Path(outroot, f'{t}xHSC/')
    for f in tracerroot.glob('*b2x1*'):
        print(f'Processing {f}')
        if not f.is_file():
            continue
        outname = tracerroot / f.name.replace('b2x1', 'b2x3')
        print(f'Renaming {f} to {outname}')
        Path.rename(f, outname)
```
