```python
import matplotlib.pyplot as plt
svar = True

tgt_list = ['BGS_ANY', 'ELGnotqso', 'LRG', 'QSO', 'HSC']
bins_list = [file_reader.get_bins(tgt) for tgt in tgt_list]  # Matching bins to targets

fig, axs = plt.subplots(2, 3, figsize=(10, 8))
axs = axs.flatten()  # Flatten for easy indexing
cmap = plt.get_cmap('plasma')

for idx, (tgt, bins) in enumerate(zip(tgt_list, bins_list), start=1):
    ax = axs[idx-1]
    avg_z = []
    for a in range(1, len(bins)):
        try:
            result = TwoPointCorrelationFunction.load(file_reader.get_auto_file(a, 0, tgt))
            if svar:
                s, xiell = result(ells=(0, 2), return_sep=True)
            else:
                s = result.sep
        except FileNotFoundError:
            continue
        mid_z = (bins[a-1] + bins[a]) / 2
        ax.plot(
            s,
            (s)**(2 if svar else 1) * (xiell[0] if svar else result.corr), 
            #'o',
            markersize=2,
            label=f'){bins[a-1]:.1f},{bins[a]:.1f}]',
            color=cmap(a / len(bins)),
            alpha=0.6,
            )
        avg_z.append(mid_z)

    avg_z = np.array(avg_z)
    mean_z = np.mean(avg_z)
    
    ax.set_title(tgt)
    ax.legend(fontsize=6, ncols=2)
    if svar:
        ax.set_ylabel(r'$s^2\xi(s)$')
        ax.set_xlabel(r'$s$ [Mpc/h]')
        ax.set_xlim(0, 200)
    else:
        ax.set_xlabel(r'$\theta$ [arcsec]')
        ax.set_xlabel(r'$\theta$ [deg]')
        sec_ax = ax.secondary_xaxis('top', functions=(
            lambda x, z=mean_z: ct.arcsec2hMpc(x*3600, z=z),
            lambda x, z=mean_z: ct.hMpc2arcsec(x/3600, z=z)
        ))
        sec_ax.set_xlabel(r'x $[h^{-1}Mpc]$'  + f' at z={mean_z:.2f}')
    ax.grid()

fig.delaxes(axs[len(tgt_list)])
plt.tight_layout()
```

```python
ROOT = Path('../autos/auto_rppi_desi')
file_reader = cu.CorrFileReader(ROOT)
import matplotlib.pyplot as plt
rppi = True

tgt_list = ['BGS_ANY', 'ELGnotqso', 'LRG', 'QSO', 'HSC']
bins_list = [file_reader.get_bins(tgt) for tgt in tgt_list] # Matching bins to targets

fig, axs = plt.subplots(2, 3, figsize=(10, 8))
axs = axs.flatten()  # Flatten for easy indexing
cmap = plt.get_cmap('plasma')

for idx, (tgt, bins) in enumerate(zip(tgt_list, bins_list), start=1):
    ax = axs[idx-1]
    avg_z = []
    for a in range(1, len(bins)):
        try:
            result = TwoPointCorrelationFunction.load(file_reader.get_auto_file(a, 0, tgt))
            if rppi:
                s, wp = result(pimax=None, return_sep=True)
            else:
                s = result.sep
        except FileNotFoundError:
            continue
        mid_z = (bins[a-1] + bins[a]) / 2
        ax.plot(
            s,
            s*wp if rppi else result.corr, 
            #'o',
            markersize=2,
            label=f'){bins[a-1]:.1f},{bins[a]:.1f}]',
            color=cmap(a / len(bins)),
            alpha=0.6,
            )
        avg_z.append(mid_z)

    avg_z = np.array(avg_z)
    mean_z = np.mean(avg_z)
    
    ax.set_title(tgt)
    ax.legend(fontsize=6, ncols=2)
    if rppi:
        ax.set_ylabel(r'$r_{p} w_{p}(r_{p})$')
        ax.set_xlabel(r'$r_{p}$')
        ax.set_xlim(0, 200)
    else:
        ax.set_xlabel(r'$\theta$ [arcsec]')
        ax.set_xlabel(r'$\theta$ [deg]')
        sec_ax = ax.secondary_xaxis('top', functions=(
            lambda x, z=mean_z: ct.arcsec2hMpc(x*3600, z=z),
            lambda x, z=mean_z: ct.hMpc2arcsec(x/3600, z=z)
        ))
        sec_ax.set_xlabel(r'x $[h^{-1}Mpc]$'  + f' at z={mean_z:.2f}')
    ax.grid()

fig.delaxes(axs[len(tgt_list)])
plt.tight_layout()
```
