```python
import numpy as np
import pandas as pd
import importlib
import matplotlib.pyplot as plt

from matplotlib.lines import Line2D
from matplotlib.patches import Patch
from matplotlib.ticker import MultipleLocator
from pathlib import Path
from scipy.integrate import simpson

import src.statistics.spline as spline
import src.analysis.plots as plots
import src.statistics.combination as comb
import src.statistics.corrfiles as cf

ROOT = cf.get_base_dir()
```

```python
bins_hsc = np.arange(0.3, 1.8, 0.3)

markers = ["x", "s", "^", "D"]
linestyles = ["-", "--", "-.", ":"]
colormaps = [plt.get_cmap(c) for c in ["Reds", "Blues", "Greens", "Purples"]]
colors = [cmap(0.8) for cmap in colormaps]
tracers = ["BGS_ANY", "LRG", "ELGnotqso", "QSO"]
tracer_trad = {
    "BGS_ANY": "BGS",
    "LRG": "LRG",
    "ELGnotqso": "ELG",
    "QSO": "QSO",
}

PAPER_FIGURES_ROOT = Path(
    "/global/cfs/projectdirs/desi/users/jchdj/desi-y3-hsc/paper/figures/"
)

# plot infrastructure
pm = plots.PlotManager(root=PAPER_FIGURES_ROOT, overwrite=True)
```

```python
scale_cut = [0.3, 3]
version = "vF"
```

```python
importlib.reload(spline)
s_data = {}
expect = {}
show_knots = False
z_all = np.linspace(0, 3, 1000)
names = ["npz_cross", "npz_bs", "npz_bs_bp", "npz_bs_bp_mag"]
name_to_label = {
    "npz_cross": "No bias evolution",
    "npz_bs": r"$b_s(z)$",
    "npz_bs_bp": r"$b_s(z), b_p(z)$",
    "npz_bs_bp_mag": r"$b_s(z), b_p(z), \mu$",
}

for tomo in [1, 2, 3, 4]:
    spline_dict = {
        name: spline.BayesianBSpline.from_saved_model(
            f"{ROOT}results/splines_{scale_cut[0]}_{scale_cut[1]}_{version}/spl_{name}_{tomo}"
        )
        for name in names
    }

    with pm.make_plot(f"nz_spline_tomo{tomo}", figsize=(9, 3.6), show=True) as (
        fig,
        ax_main,
    ):
        cmap = plt.get_cmap("plasma_r")
        markers = ["x", "s", "^", "D"]
        linestyles = ["-", "--", "-.", ":"]

        for ind, name in enumerate(names):
            spl_study = spline_dict[name]

            mask = (z_all <= spl_study.zv.max()) & (z_all >= spl_study.zv.min())
            z_eval = z_all[mask]

            n_eval_points = 200
            nz_samples = spl_study.get_samples(
                z_eval=z_eval, n_eval_points=n_eval_points
            )
            basis_eval = spl_study._create_evaluation_basis(z_eval)

            nz_median = np.percentile(nz_samples, 50, axis=0)
            norm_median = simpson(nz_median, z_eval)
            nz_samples_norm = nz_samples / simpson(nz_samples, z_eval, axis=1)[:, None]

            nz_median = np.percentile(nz_samples_norm, 50, axis=0)
            nz_mean = np.mean(nz_samples_norm, axis=0)
            nz_std = np.std(nz_samples_norm, axis=0)
            nz_lower = np.percentile(nz_samples_norm, 16, axis=0)
            nz_upper = np.percentile(nz_samples_norm, 84, axis=0)

            color = cmap((ind + 0.5) / (len(names)))

            nz_median_all = np.zeros(len(z_all))
            nz_mean_all = np.zeros(len(z_all))
            nz_mean_all[mask] = nz_mean
            nz_median_all[mask] = nz_median
            nz_lower_all = np.zeros(len(z_all))
            nz_lower_all[mask] = nz_lower
            nz_upper_all = np.zeros(len(z_all))
            nz_upper_all[mask] = nz_upper

            s_data[f"z"] = z_all
            if (name == "npz_bs_bp") or (name == "npz_bs_bp_mag"):
                if "mag" in name:
                    add = "_mu"
                else:
                    add = ""
                s_data[f"{tomo}/nz_median{add}"] = nz_median_all
                s_data[f"{tomo}/nz_mean{add}"] = nz_mean_all
                s_data[f"{tomo}/nz_lower{add}"] = nz_lower_all
                s_data[f"{tomo}/nz_upper{add}"] = nz_upper_all

            ax_main.errorbar(
                spl_study.zv + (0.005 * (ind - 1)),
                spl_study.nz / norm_median,
                yerr=spl_study.nz_err / norm_median,
                fmt=markers[ind],
                color=color,
                alpha=1,
                capsize=3.5,
                capthick=1.5,
                label=name_to_label[name],
            )
            ax_main.plot(
                z_eval,
                nz_median,
                color=color,
                alpha=0.8,
                linewidth=2,
                linestyle=linestyles[ind],
            )
            ax_main.fill_between(
                z_eval,
                nz_lower,
                nz_upper,
                color=color,
                alpha=0.3,
            )
            print("\n")
            print("Expectation")
            expect_samples = [
                np.trapezoid(nzsn * z_eval, z_eval) for nzsn in nz_samples_norm
            ]
            percentiles = np.percentile(expect_samples, [16, 50, 84])
            print(f"16th, 50th, 84th percentiles: {percentiles}")
            expect_nz = np.trapezoid(nz_mean * z_eval, z_eval)
            print(f"Expectation from mean n(z): {expect_nz}")
            print("\n")
            expect[f"{tomo}/{name}"] = percentiles
            expect[f"{tomo}/{name}_mean"] = expect_nz

        ax_main.legend(fontsize=12.5, loc="upper left", framealpha=0.9)
        ax_main.text(
            0.82,
            0.95,
            f"Bin {tomo}\nz:{0.3*tomo:.1f}-{0.3*(tomo+1):.1f}",
            transform=ax_main.transAxes,
            verticalalignment="top",
            bbox=dict(facecolor="white", alpha=0.8, edgecolor="none"),
            fontsize=16,
        )

        if show_knots:
            knot_positions = spl_study.knots[spl_study.degree : -spl_study.degree]
            for i, knot in enumerate(knot_positions):
                alpha_val = 0.6 if i == 0 else 0.4
                label_val = "Knots" if i == 0 else None
                ax_main.axvline(
                    knot,
                    color="gray",
                    linestyle="--",
                    alpha=alpha_val,
                    linewidth=1,
                    label=label_val,
                )

        ax_main.set_xlabel("Redshift (z)", fontsize=16)
        ax_main.set_ylabel(r"$n(z)$", fontsize=16)
        ax_main.axhline(0, color="black", linestyle="--")
        ax_main.grid(True, alpha=0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

tomo = 2
spl = spline.BayesianBSpline.from_saved_model(
    f"{ROOT}results/splines_{scale_cut[0]}_{scale_cut[1]}_{version}/spl_npz_bs_bp_mag_{tomo}"
)
n_eval_points = 600

with pm.make_plot(
    f"spline_showcase",
    figsize=(10, 8),
    show=True,
    tight_layout=False,
    custom_layout=True,
) as fig:
    z_eval = np.linspace(spl.zv.min(), spl.zv.max(), n_eval_points)
    nz_samples = spl.get_samples(z_eval=z_eval, n_eval_points=n_eval_points)
    basis_eval = spl._create_evaluation_basis(z_eval)

    nz_median = np.percentile(nz_samples, 50, axis=0)
    nz_lower = np.percentile(nz_samples, 16, axis=0)
    nz_upper = np.percentile(nz_samples, 84, axis=0)
    coeffs_samples = spl.coeffs_samples

    # Grid spec with reduced top height
    gs = fig.add_gridspec(2, 2, height_ratios=[1.3, 1], hspace=0.35, wspace=0.25)
    ax_main = fig.add_subplot(gs[0, :])

    info_text = f"Knots: {spl.n_knots}\nSpline degree: {spl.degree}"
    coeffs_nnls, _ = spl._get_init(spl.nz)
    nnls_pred_eval = basis_eval @ coeffs_nnls

    # Random posterior samples for visual uncertainty
    n_show_samples = 800
    col_samples = "indigo"
    rand_idx = np.random.choice(len(nz_samples), n_show_samples, replace=False)
    for idx in rand_idx:
        ax_main.plot(
            z_eval,
            nz_samples[idx, :],
            color=col_samples,
            alpha=0.12,
            linewidth=0.13,
            zorder=1,
        )

    # add a label for the samples in the legend
    # set x outside the xlims
    ax_main.axhline(150, color=col_samples, lw=2, alpha=0.5, label="Samples")

    ax_main.errorbar(
        spl.zv,
        spl.nz,
        yerr=spl.nz_err,
        fmt="s",
        color="orange",
        alpha=1,
        capsize=4,
        capthick=1.5,
        label=r"n($z$)",
    )

    knot_positions = spl.knots[spl.degree : -spl.degree]
    for i, knot in enumerate(knot_positions):
        ax_main.axvline(
            knot,
            color="darkgray",
            linestyle="--",
            alpha=0.7,
            linewidth=1,
            label="Knots" if i == 0 else None,
        )

    ax_main.plot(z_eval, nz_median, color="red", linewidth=2, label="Posterior median")
    ax_main.fill_between(z_eval, nz_lower, nz_upper, color="red", alpha=0.3, hatch="//")

    ax_main.axhline(0, color="black", linestyle="--", linewidth=1)
    ax_main.set_xlabel(r"Redshift ($z$)", fontsize=14)
    ax_main.set_ylabel(r"n($z$)", fontsize=14)
    ax_main.text(
        0.02,
        0.96,
        info_text,
        transform=ax_main.transAxes,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.8),
        fontsize=14,
    )
    ax_main.plot(
        z_eval,
        nnls_pred_eval,
        color="navy",
        linewidth=3,
        linestyle=":",
        alpha=1,
        label="NNLS",
        zorder=5,
    )
    ax_main.grid(True, alpha=0.3)
    ax_main.legend(fontsize=12, loc="upper right", framealpha=0.9)
    ax_main.set_ylim(-0.25 * spl.nz.max(), 1.3 * spl.nz.max())

    ax_basis = fig.add_subplot(gs[1, 0])
    n_show = spl.n_basis
    colors = plt.get_cmap("plasma")(np.linspace(0, 1, n_show))

    for i in range(n_show):
        basis_func = basis_eval[:, i]
        ax_basis.plot(z_eval, basis_func, color=colors[i], alpha=1, linewidth=2.1)

    ax_basis.axhline(0, color="black", linewidth=2)

    ax_basis.set_xlabel(r"Redshift ($z$)", fontsize=14)
    ax_basis.set_ylabel("Amplitude", fontsize=14)
    ax_basis.set_title("B-spline basis functions", fontsize=14)
    ax_basis.grid(True, alpha=0.3)

    # --- Right column: coefficients + Dirichlet prior using nested GridSpec ---
    gs_right = gs[1, 1].subgridspec(2, 1, height_ratios=[0.15, 0.3], hspace=0.05)

    ax_coeff = fig.add_subplot(gs_right[1, 0])

    ax_dirichlet = fig.add_subplot(gs_right[0, 0], sharex=ax_coeff)

    prior_concentration = 3
    base_alpha = 0.05
    coeffs_init, init_A = spl._get_init(spl.nz)
    dirichlet_alpha = base_alpha + prior_concentration * coeffs_init

    x_pos = np.arange(spl.n_basis)
    ax_dirichlet.fill_between(x_pos, 0, dirichlet_alpha, color="crimson", alpha=0.6)
    ax_dirichlet.set_ylabel(r"$\alpha_i$", fontsize=12)
    ax_dirichlet.grid(True, alpha=0.3)
    ax_dirichlet.text(
        0.02,
        0.7,
        "Dir($\\alpha$) prior",
        fontsize=12,
        ha="left",
        transform=ax_dirichlet.transAxes,
    )
    ax_dirichlet.set_xticks([])
    ax_dirichlet.set_yticks([])

    coeff_medians = np.median(coeffs_samples, axis=0)
    coeff_stds = np.std(coeffs_samples, axis=0)

    lower_errors = coeff_medians - np.maximum(coeff_medians - coeff_stds, 0)
    upper_errors = coeff_stds

    ax_coeff.bar(
        x_pos,
        coeff_medians,
        yerr=np.array([lower_errors, upper_errors]),
        capsize=3,
        alpha=0.7,
        color="orange",
        edgecolor="darkorange",
        linewidth=2,
        error_kw={"ecolor": "red", "elinewidth": 2, "capsize": 3, "capthick": 1.5},
    )
    ax_coeff.scatter(
        x_pos,
        coeff_medians,
        color="darkred",
        s=15,
        alpha=0.8,
        zorder=3,
        label=r"Median $c_i$",
        marker="d",
    )
    ax_coeff.set_xlabel("Basis function index", fontsize=14)
    ax_coeff.set_ylabel(r"$c_i$", fontsize=14)
    ax_coeff.set_title(r"Posterior $c_i$ distribution", fontsize=14)
    ax_coeff.grid(True, alpha=0.3, axis="y")
    ax_coeff.legend(fontsize=11)
    ax_coeff.set_xlim(-0.5, spl.n_basis - 0.5)
    ax_coeff.axhline(0, color="black", linestyle="--", linewidth=1)
    ax_coeff.set_ylim(-0.05, 0.4)

    if spl.n_basis > 15:
        tick_spacing = max(1, spl.n_basis // 10)
        ax_coeff.set_xticks(x_pos[::tick_spacing])
        ax_coeff.set_xticklabels(x_pos[::tick_spacing])
```
