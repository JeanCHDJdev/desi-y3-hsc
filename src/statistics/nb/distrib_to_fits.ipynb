```python
import numpy as np

from astropy.table import Table
from pathlib import Path

import src.statistics.corrfiles as cf

ROOT = cf.get_base_dir()

scale_cut = [1, 5]
file = Path(f"{ROOT}/results/distributions_{scale_cut[0]}_{scale_cut[1]}_vF.npz")
assert file.exists(), f"File {file} does not exist."

distrib = np.load(file)
```

```python
z_mid = distrib["z"]
dz = np.diff(z_mid)
z_low = np.empty_like(z_mid)
z_high = np.empty_like(z_mid)
z_low[1:] = z_mid[1:] - 0.5 * dz
z_low[0] = z_mid[0] - 0.5 * dz[0]
z_high[:-1] = z_mid[:-1] + 0.5 * dz
z_high[-1] = z_mid[-1] + 0.5 * dz[-1]
```

```python
# make a new fits table with the distributions as columns
term = "median"
out = Path(f"{ROOT}/results/distributions_{scale_cut[0]}_{scale_cut[1]}_{term}_vF.fits")
table = Table()
table["Z_MID"] = z_mid
table["Z_LOW"] = z_low
table["Z_HIGH"] = z_high

print(len(z_mid), len(z_low), len(z_high), len(distrib[f"1/nz_{term}_mu"]))

for i in [1, 2, 3, 4]:
    table[f"BIN_{i}"] = distrib[f"{i}/nz_{term}_mu"]

table.write(out, overwrite=True)
```
