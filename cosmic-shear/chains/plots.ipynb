```python
filename = "/global/cfs/projectdirs/desi/users/jchdj/cosmosis-standard-library/output-corrected-nz-real/hsc-y3-shear.txt"
with open(filename, "r") as f:
    header = f.readline().lstrip("#").strip().split()
print(f"Columns in the chain file: {header}")

df = pd.read_csv(
    filename, 
    delim_whitespace=True, 
    comment="#", 
    names=header
    )
burn_frac = 0.2  # 20% burn-in
burn = int(len(df) * burn_frac)
df_burnt = df#.iloc[burn:]

params = [
    "cosmological_parameters--omega_m",
    "COSMOLOGICAL_PARAMETERS--S_8",
    'wl_photoz_errors--bias_1',
    'wl_photoz_errors--bias_2',
    'wl_photoz_errors--bias_3',
    'wl_photoz_errors--bias_4'
]
print(df_burnt[params].describe())

samples = MCSamples(
    samples=df_burnt[params].values,
    names=params,
    labels=[r"\Omega_m", r"S_8", r"\Delta z_1", r"\Delta z_2", r"\Delta z_3", r"\Delta z_4"],
    weights=df_burnt["post"].values if "post" in df_burnt else None
)

g = gdplots.get_subplot_plotter()
g.triangle_plot([samples], params, filled=True)
for ax in g.subplots.flatten():
    if ax is not None:
        ax.grid(True, linestyle="--", alpha=0.5)
        ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=8))
        ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=8))

        ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
        ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
        ax.grid(which='minor', linestyle=':', alpha=0.3)

plt.show()
```
